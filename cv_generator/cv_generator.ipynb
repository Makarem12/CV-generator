```python
# welcoming message
print("Welcome to the CV Generator!")
print("This tool will help you create your CV based on a template.")
print("You will be prompted to enter information for various sections of your CV.")
print("Let's get started!")

def fill_template(template, user_data):
    filled_template = template.format(**user_data)
    return filled_template

template_path = "../assets/cv_template.txt"
with open(template_path, 'r') as file:
    template = file.read()

# Prompt for user input
user_data = {}
user_data['Name'] = input("Enter your name: ")
user_data['phone_number'] = input("Enter your phone number: ")
user_data['email'] = input("Enter your email: ")
user_data['city'] = input("Enter your city: ")
user_data['university_name'] = input("Enter your university name: ")
user_data['major'] = input("Enter your major: ")
user_data['graduation_date'] = input("Enter your graduation date: ")
user_data['company_name'] = input("Enter your company name: ")
user_data['job_title'] = input("Enter your job title: ")
user_data['you responsibilities'] = input("Enter summary of your key responsibilities: ")
user_data['skill-01'] = input("Enter skill 1: ")
user_data['skill-02'] = input("Enter skill 2: ")
user_data['skill-03'] = input("Enter skill 3: ")
user_data['skill-04'] = input("Enter skill 4: ")
user_data['Arabic_language_level'] = input("Enter Arabic language level: ")
user_data['English_language_level'] = input("Enter English language level: ")

# Fill the template with user data
filled_cv = fill_template(template, user_data)

# Display completed CV
print("\nCompleted CV:")
print(filled_cv)

# Write completed CV to a file
with open('cv_output.txt', 'w') as file:
    file.write(filled_cv)
```

```python
def read_template(file_path):
    try:
        with open(file_path, 'r') as file:
            template = file.read().strip()
        return template
    except FileNotFoundError:
        raise FileNotFoundError(f"File not found at path: {file_path}")

def parse_template(template):
    language_parts = []
    parsed_template = template
    # Find language parts in the template and remove them
    while '{' in parsed_template and '}' in parsed_template:
        start_index = parsed_template.find('{')
        end_index = parsed_template.find('}')
        language_parts.append(parsed_template[start_index:end_index+1])
        parsed_template = parsed_template[:start_index] + parsed_template[end_index+1:]
    return parsed_template.strip(), tuple(language_parts)

def merge(template, language_parts):
    merged_text = template
    for part in language_parts:
        merged_text = merged_text.replace(part, input(f"Enter value for {part}: "))
    return merged_text

# Testing the functions
def test_read_template():
    # Valid file path
    file_path = 'assets/cv_template.txt'
    assert read_template(file_path) == "I'm {Name} and I have Bachelor's degree in {major}."

    # Invalid file path
    invalid_file_path = 'invalid_path.txt'
    try:
        read_template(invalid_file_path)
    except FileNotFoundError as e:
        assert str(e) == f"File not found at path: {invalid_file_path}"

def test_parse_template():
    template = "I'm {Name} and I have Bachelor's degree in {major}."
    parsed_template, language_parts = parse_template(template)
    assert parsed_template == "I'm  and I have Bachelor's degree in ."
    assert language_parts == ('{Name}', '{major}')

def test_merge():
    template = "I'm {Name} and I have Bachelor's degree in {major}."
    language_parts = ('{Name}', '{major}')
    merged_text = merge(template, language_parts)
    assert merged_text == "I'm John and I have Bachelor's degree in Computer Science."

# Run tests
if __name__ == "__main__":
    test_read_template()
    test_parse_template()
    test_merge()
    print("All tests passed!")
```
